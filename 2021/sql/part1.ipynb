```python
from datetime import datetime
import math
```

```python
def date_manipulation(dates):
    
    dates_ = [datetime.strptime(x, '%Y-%m-%d') for x in dates]
    start = min(dates_)
    dates_ = sorted(dates_)
    return [math.ceil((x - start).days/7)  for x in dates_]
    
    

date_manipulation(['2021-09-01', '2021-06-18', '2021-05-14', '2021-06-11', 
 '2021-09-25', '2021-09-08', '2021-05-02', '2021-04-30', '2021-07-08'])
```
