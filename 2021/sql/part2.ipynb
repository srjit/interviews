```python
import pandas as pd
```

```python
df = pd.read_csv("sales-rec.csv")
df.head()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"]).apply(lambda x: str(x)[:10])
df["Ship Date"] = pd.to_datetime(df["Ship Date"]).apply(lambda x: str(x)[:10])
```

```python
df.to_csv("sales.csv", index=None)
```
