```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("Advertising.csv")
data.head()
```

```python
reg = LinearRegression()
reg.fit(data["Radio"].values.reshape(-1,1), data["Sales"])
reg
```

```python
plt.scatter(data["Radio"], data["Sales"])
plt.plot(data["Radio"], [reg.intercept_ + reg.coef_[0] * x for x in data["Radio"].values], color = "k", ls="--")

plt.xlabel("Radio")
plt.ylabel("Sales")
plt.show()
```
