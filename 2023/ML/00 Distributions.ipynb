```python
import numpy as np
import matplotlib.pyplot as plt
import viz
```

```python
class Uniform:

    def __init__(self, a: int, b: int):
        self.a = a
        self.b = b

    def pdf(self, x: int) -> float:
        if self.a <= x <= self.b:
            return 1 / (self.b - self.a + 1)
        else:
            return 0
```

```python
uniform = Uniform(3, 6)
xvalues = [round(x,1) for x in np.linspace(1.0, 9.0, 80)]
yvalues = [uniform.pdf(x) for x in xvalues]
```

```python
# for x,y in zip(xvalues[18:54], yvalues[18:54]):
#     print((x, y))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
viz.plot(xvalues, 
         yvalues, 
         ax=ax, 
         title="Uniform Distribution",
         xlabel="X", 
         marker="o",
         ylabel="Y", 
         format_y=False, 
         make_x_string=False)
plt.show()
```
